```python
import re
```

```python
msg = "ThisIsATest"
```

```python
re.findall(r"[A-Z][a-z]*", msg)
```

```python
splitter = re.compile(r"[A-Z][a-z]*").findall
splitter(msg)
```

```python
table["Type"].apply(splitter)
```

```python
table.assign(bias=1)
```

```python
t2 = table.assign(Type=table["Type"].apply(splitter))
t2
```

```python
from functools import reduce
```

```python
reduce(set.union, t2["Type"].apply(set))
```

```python
# This is what "reduce" does:
result = set()
for item in t2["Type"].apply(set):
    result = set.union(result, item)

result
```
